```python
def load_spacy_model():
    try:
        return spacy.load("de_core_news_sm")
    except IOError:
        print("Downloading German language model...")
        spacy.cli.download("de_core_news_sm")
        return spacy.load("de_core_news_sm")

def marburg_ner(text):
    nlp = load_spacy_model()
    doc = nlp(text)

    marburg_related = []

    # Define Marburg-related patterns
    marburg_patterns = [
        r'\bMarburg\b',
        r'\bMengsberg\b',
        r'\bMarburg-Biedenkopf\b',
        r'\bLandkreis Marburg\b',
        r'\bPhilipps-Universität\b'
    ]

    # Check for Marburg-related patterns
    for pattern in marburg_patterns:
        matches = re.finditer(pattern, text, re.IGNORECASE)
        for match in matches:
            marburg_related.append((match.group(), match.start(), match.end(), "MARBURG_RELATED"))

    # Add named entities from spaCy
    for ent in doc.ents:
        if ent.label_ in ["LOC", "ORG"]:
            marburg_related.append((ent.text, ent.start_char, ent.end_char, ent.label_))

    return sorted(marburg_related, key=lambda x: x[1])

# Example usage
# text = "\n".join([
#     'Lauenhain (Kreis Hainichen), Fahnenträger eines internationalen Zeltlagers',
#     'Fenster (Bauelement)',
#     'Kassel, Infanteriekasernen an der Königsstraße, Bestandsaufnahme, Situationsplan',
#     'Waldeck, Domäne, Wohnhaus, Umbauentwurf, Aufrisse',
#     'Kassel, Opernhaus (Altes Hoftheater), Aborte im Flügelgebäude, Entwurf, Grundriss',
#     'Mengsberg, Oberförster-Etablissement, Bauaufnahmen, Grundrisse, Aufrisse, Querschnitte, Längsschnitte und Lageplan',
#     'Schafherde in der Eifel'
# ])

# results = marburg_ner(text)
# for entity, start, end, label in results:
#     print(f"{entity} ({start}:{end}) - {label}")
```

```python
import requests
import re

def query_osm_for_marburg():
    overpass_url = "http://overpass-api.de/api/interpreter"
    overpass_query = """
    [out:json];
    area["name"="Marburg"]["admin_level"="8"]->.searchArea;
    (
      node["place"](area.searchArea);
      way["place"](area.searchArea);
      relation["place"](area.searchArea);
      node["historic"](area.searchArea);
      way["historic"](area.searchArea);
      relation["historic"](area.searchArea);
    );
    out body;
    >;
    out skel qt;
    """

    response = requests.get(overpass_url, params={'data': overpass_query})
    return response.json()

def create_regex_patterns(osm_data):
    patterns = {
        'MARBURG_DISTRICT': [],
        'MARBURG_LANDMARK': [],
        'MARBURG_STREET': []
    }

    for element in osm_data['elements']:
        if 'tags' in element:
            name = element['tags'].get('name')
            if name:
                if element['tags'].get('place') in ['suburb', 'quarter']:
                    patterns['MARBURG_DISTRICT'].append(r'\b' + re.escape(name) + r'\b')
                elif element['tags'].get('historic') or element['tags'].get('tourism'):
                    patterns['MARBURG_LANDMARK'].append(r'\b' + re.escape(name) + r'\b')
                elif element['tags'].get('highway'):
                    patterns['MARBURG_STREET'].append(r'\b' + re.escape(name) + r'\b')

    return patterns

def marburg_ner(text, patterns):
    entities = []
    for category, regex_list in patterns.items():
        for regex in regex_list:
            matches = re.finditer(regex, text)
            for match in matches:
                entities.append((match.group(), match.start(), match.end(), category))
    return sorted(entities, key=lambda x: x[1])

# Example usage
osm_data = query_osm_for_marburg()
regex_patterns = create_regex_patterns(osm_data)

sample_text = "The Elisabethkirche in Marburg's Altstadt is a famous landmark. Many students live in the Südviertel district."
results = marburg_ner(sample_text, regex_patterns)

for entity, start, end, label in results:
    print(f"{entity} ({start}:{end}) - {label}")
```

```python
import torch

def setup_german_ner_model():
    model_name = "flair/ner-german-large"
    tokenizer = AutoTokenizer.from_pretrained("bert-base-german-cased")
    model = AutoModelForTokenClassification.from_pretrained(model_name)

    # Define a custom pipeline that uses Flair's label scheme
    def flair_ner_pipeline(text):
        inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
        with torch.no_grad():
            outputs = model(**inputs)

        predictions = torch.argmax(outputs.logits, dim=2)
        tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])

        entities = []
        current_entity = None
        for token, pred in zip(tokens, predictions[0]):
            if pred != 0:  # 0 is usually the 'O' (Outside) label
                label = model.config.id2label[pred.item()]
                if current_entity and current_entity['entity_group'] != label:
                    entities.append(current_entity)
                    current_entity = None
                if not current_entity:
                    current_entity = {'word': token, 'entity_group': label, 'start': len(entities)}
                else:
                    current_entity['word'] += f" {token}"
            elif current_entity:
                entities.append(current_entity)
                current_entity = None

        if current_entity:
            entities.append(current_entity)

        return entities

    return flair_ner_pipeline

ner_pipeline = setup_german_ner_model()
print("German NER model (Flair) setup complete.")
```

```python
def process_german_captions(captions, ner_pipeline):
    results = {}
    for caption in captions:
        entities = ner_pipeline(caption)
        # We'll keep all entity types for now
        results[caption] = entities
    return results
```
