```python
# prompt: read this json file /content/drive/MyDrive/time_machine_project/buildings_dict.json

import json
import os

# Check if the file exists
file_path = "/content/drive/MyDrive/time_machine_project/buildings_dict.json"
if os.path.exists(file_path):
  with open(file_path, 'r') as f:
    try:
      data = json.load(f)
      print(data)
    except json.JSONDecodeError as e:
      print(f"Error decoding JSON: {e}")
else:
  print(f"File not found: {file_path}")
```

```python
import pandas as pd

building_data = []
for region, streets in data.items():
    for street_name, buildings in streets.items():
        for building in buildings:
            conc_name_add = f"{building['name']}, {building['address']}"  # Concatenated field

            building_data.append({
                "region": region,
                "street": street_name,
                "name": building["name"],
                "address": building["address"],
                "conc_name_add": conc_name_add,
                "id_link": building.get("id_link", None),
                "address_code": building.get("address_code", None),
                "gebnum": building.get("gebnum", None)
            })

df_buildings = pd.DataFrame(building_data)

```

```python
folder_path = "/content/drive/MyDrive/time_machine_project/"
file_name = "df_buildings.csv"

# Ensure the directory exists
os.makedirs(folder_path, exist_ok=True)

# Save DataFrame (replace df with your actual DataFrame)
df_buildings.to_csv(os.path.join(folder_path, file_name), index=False)

print(f"DataFrame saved at: {folder_path}{file_name}")
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
# Load SBERT model (use a lightweight model like 'all-MiniLM-L6-v2' for efficiency)
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Ensure the column is string type and handle missing values
df_buildings['conc_name_add'] = df_buildings['conc_name_add'].astype(str).fillna('')
```

```python
# Encode column to embeddings
embeddings = model.encode(df_buildings['conc_name_add'].tolist(), show_progress_bar=True)
```

```python
# Convert embeddings into separate columns
embedding_dim = embeddings.shape[1]  # Get the number of dimensions
embedding_cols = [f"conc_name_emb_{i}" for i in range(embedding_dim)]
```

```python
# Add embeddings to DataFrame
df_embeddings = pd.DataFrame(embeddings, columns=embedding_cols)
```

```python
df_embeddings
```

```python
df_vecs_buildings = pd.concat([df_buildings, df_embeddings], axis=1)

print("Vectorization complete. DataFrame df_vecs_buildings now contains SBERT embeddings.")
```

```python
folder_path = "/content/drive/MyDrive/time_machine_project/"
file_name = "df_vecs_buildings.csv"

# Ensure the directory exists
os.makedirs(folder_path, exist_ok=True)

# Save DataFrame (replace df with your actual DataFrame)
df_vecs_buildings.to_csv(os.path.join(folder_path, file_name), index=False)

print(f"DataFrame saved at: {folder_path}{file_name}")
```
