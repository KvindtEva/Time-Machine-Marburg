```python
df_cluster_with_text_truncated = df_cluster_with_text[['Label', 'key_to_cluster']]
```

```python
df_cluster_with_text_truncated
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Load the same SBERT model used for "conc_name_add"
model = SentenceTransformer('all-MiniLM-L6-v2')

# Convert "Labels" column to SBERT embeddings
label_embeddings = model.encode(df_cluster_with_text_truncated['Label'].tolist(), normalize_embeddings=True)
```

```python
label_embeddings
```

```python
embedding_dim =label_embeddings.shape[1]
embedding_cols = [f"conc_name_emb_{i}" for i in range(embedding_dim)]
```

```python
df_label_embeddings = pd.DataFrame(label_embeddings, columns=embedding_cols)
```

```python
df_label_embeddings
```

```python
df_cluster_with_text_truncated_aligend = df_cluster_with_text_truncated.reset_index(drop=True)
```

```python
df_cluster_with_text_truncated_aligend
```

```python
df_cluster_with_text_truncated_new = pd.concat([df_cluster_with_text_truncated_aligend, df_label_embeddings], axis=1)
```
